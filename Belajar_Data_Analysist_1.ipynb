```python
import pandas as pd

data = {
    'Nama': ['Zall', 'Budi', 'Siti', 'Andi', 'Lina'],
    'Umur': [20, 22, 19, 21, 20],
    'Jurusan': ['Informatika', 'Sistem Informasi', 'Informatika', 'Teknik Komputer', 'Informatika'],
    'IPK': [3.75, 3.20, 3.90, 2.95, 3.85],
    'Kota': ['Jakarta', 'Bandung', 'Surabaya', 'Medan', 'Jakarta']
}

df = pd.DataFrame(data)
df
```

```python
# 1. Informasi dasar
print("üìä Bentuk ", df.shape)
print("\nüìã Info kolom:")
df.info()

# 2. Statistik deskriptif
print("\nüìà Statistik IPK & Umur:")
df.describe()

# 3. Cek missing value
print("\n‚ùì Missing value per kolom:")
df.isnull().sum()

# 4. Analisis per kelompok
print("\nüéì Rata-rata IPK per Jurusan:")
df.groupby('Jurusan')['IPK'].mean().sort_values(ascending=False)

# 5. Visualisasi
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Jurusan', y='IPK')
plt.title('Distribusi IPK per Jurusan')
plt.xticks(rotation=15)
plt.show()
```

```python
# Misal: simpan ringkasan IPK per jurusan
ringkasan = df.groupby('Jurusan')['IPK'].agg(['mean', 'count']).round(2)
ringkasan.to_csv('ringkasan_ipk.csv')

# Download ke laptop
from google.colab import files
files.download('ringkasan_ipk.csv')
```
